**Multivariate Analysis (Higher-Dimensional Data Relationships)**<br>
Objective: Analyze relationships in higher-dimensional data using pair plots and PCA.

Title: Pair Plots<br>

Task 1: Create a pair plot for the Iris dataset.<br>
Task 2: Focus on a subset of features (e.g., only petal dimensions).<br>
Task 3: Exclude one class to observe differences.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load Iris dataset with target labels
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Task 1: Pair plot for the full Iris dataset
sns.pairplot(df, hue='species', diag_kind='kde', palette='bright')
plt.suptitle('Pair Plot: All Features with Species', y=1.02)
plt.show()

# Task 2: Pair plot focusing only on petal length and petal width
sns.pairplot(df, vars=['petal length (cm)', 'petal width (cm)'], hue='species', diag_kind='kde', palette='bright')
plt.suptitle('Pair Plot: Petal Dimensions with Species', y=1.02)
plt.show()

# Task 3: Exclude one class (e.g., 'setosa') and create a pair plot
df_subset = df[df['species'] != 'setosa']
sns.pairplot(df_subset, hue='species', diag_kind='kde', palette='dark')
plt.suptitle('Pair Plot: Excluding Setosa', y=1.02)
plt.show()
```

Title: Principal Component Analysis (PCA)<br>

Task 1: Perform PCA and plot the first two principal components.<br>
Task 2: Visualize explained variance by each component.<br>
Task 3: Retain more components and visualize in 3D.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
target = iris.target
target_names = iris.target_names

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# Task 1: Perform PCA and plot the first two principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[target == i, 0], X_pca[target == i, 1], label=target_name)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Principal Components')
plt.legend()
plt.grid(True)
plt.show()

# Task 2: Visualize explained variance by each component
pca_full = PCA()
pca_full.fit(X_scaled)
explained_var = pca_full.explained_variance_ratio_

plt.figure(figsize=(8,5))
plt.bar(range(1, len(explained_var)+1), explained_var, alpha=0.7, color='teal')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Each Principal Component')
plt.xticks(range(1, len(explained_var)+1))
plt.show()

# Task 3: Retain 3 components and visualize in 3D
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')
colors = ['r', 'g', 'b']
for i, target_name in enumerate(target_names):
    ax.scatter(
        X_pca_3d[target == i, 0],
        X_pca_3d[target == i, 1],
        X_pca_3d[target == i, 2],
        label=target_name,
        color=colors[i],
        edgecolor='k',
        s=60
    )
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
ax.set_title('3D PCA Visualization')
ax.legend()
plt.show()
```

**Statistical Analysis in EDA**<br>

Objective: Calculate basic statistical metrics and explore the relationship between features using correlation and covariance.<br>

Title: Descriptive Statistics<br>

Task 1: Calculate mean, median, and standard deviation of petal length.<br>
Task 2: Calculate skewness and kurtosis of sepal width.<br>
Task 3: Calculate mean, median, and standard deviation of sepal length.

```python
import pandas as pd
from sklearn.datasets import load_iris
from scipy.stats import skew, kurtosis

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Task 1: Mean, median, std of petal length
petal_length = df['petal length (cm)']
print("Petal Length:")
print(f" Mean: {petal_length.mean():.3f}")
print(f" Median: {petal_length.median():.3f}")
print(f" Standard Deviation: {petal_length.std():.3f}\n")

# Task 2: Skewness and kurtosis of sepal width
sepal_width = df['sepal width (cm)']
print("Sepal Width:")
print(f" Skewness: {skew(sepal_width):.3f}")
print(f" Kurtosis: {kurtosis(sepal_width):.3f}\n")

# Task 3: Mean, median, std of sepal length
sepal_length = df['sepal length (cm)']
print("Sepal Length:")
print(f" Mean: {sepal_length.mean():.3f}")
print(f" Median: {sepal_length.median():.3f}")
print(f" Standard Deviation: {sepal_length.std():.3f}")
```

Title: Correlation & Covariance<br>

Task 1: Compute correlation between sepal length and petal length.<br>
Task 2: Compute covariance between petal width and sepal width.<br>
Task 3: Determine the most correlated pair of features.<br>

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Task 1: Correlation between sepal length and petal length
corr_sepal_petal_length = df['sepal length (cm)'].corr(df['petal length (cm)'])
print(f"Correlation between Sepal Length and Petal Length: {corr_sepal_petal_length:.3f}")

# Task 2: Covariance between petal width and sepal width
cov_petal_sepal_width = df['petal width (cm)'].cov(df['sepal width (cm)'])
print(f"Covariance between Petal Width and Sepal Width: {cov_petal_sepal_width:.3f}")

# Task 3: Find the most correlated pair of features
corr_matrix = df.corr().abs()  # absolute correlations
# Set diagonal to 0 to ignore self-correlation
for i in range(len(corr_matrix)):
    corr_matrix.iat[i,i] = 0
max_corr = corr_matrix.unstack().idxmax()
max_corr_value = corr_matrix.unstack().max()
print(f"Most correlated pair: {max_corr} with correlation {max_corr_value:.3f}")
```
