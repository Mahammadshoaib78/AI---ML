```python
# 1. Clustering

import numpy as np
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Task 1: K-means clustering - customer segmentation (synthetic example)
customer_data = np.array([[5, 2], [10, 8], [6, 3], [9, 7], [2, 1]])
kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(customer_data)
print("K-means clusters:", clusters)

# Task 2: Hierarchical clustering - organize species (using Iris dataset)
iris = load_iris()
species_data = iris.data
linked = linkage(species_data, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(linked, labels=iris.target)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()

# Task 3: Clustering web pages (example with TF-IDF vectors and KMeans)
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "machine learning algorithms",
    "deep learning in neural networks",
    "search engine optimization techniques",
    "machine learning applications",
    "optimization of web search results"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)
kmeans_web = KMeans(n_clusters=2, random_state=0)
web_clusters = kmeans_web.fit_predict(X)
print("Web page clusters:", web_clusters)


# 2. Dimensionality Reduction

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Task 1: PCA - facial recognition dataset (using sklearn digits)
from sklearn.datasets import load_digits
digits = load_digits()
pca = PCA(n_components=2)
digits_pca = pca.fit_transform(digits.data)
plt.scatter(digits_pca[:,0], digits_pca[:,1], c=digits.target, cmap='viridis', s=15)
plt.title("PCA on Digits Dataset")
plt.show()

# Task 2: t-SNE - visualization of high-dimensional data (digits dataset again)
tsne = TSNE(n_components=2, random_state=0)
digits_tsne = tsne.fit_transform(digits.data)
plt.scatter(digits_tsne[:,0], digits_tsne[:,1], c=digits.target, cmap='viridis', s=15)
plt.title("t-SNE on Digits Dataset")
plt.show()

# Task 3: LDA - reduce variables in stock market prediction datasets (synthetic example)
# LDA requires labeled classes, so this is a supervised dimensionality reduction
X_stock = np.random.rand(100, 10)
y_stock = np.random.choice([0, 1], size=100)
lda = LDA(n_components=1)
X_lda = lda.fit_transform(X_stock, y_stock)
print("LDA transformed shape:", X_lda.shape)


# 3. Association Rule Learning

from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd

# Task 1: Market basket analysis
basket = pd.DataFrame({
    'bread': [1,0,1,1],
    'milk': [1,1,0,1],
    'butter': [0,1,1,1],
    'jam': [1,0,1,0]
})
frequent_itemsets = apriori(basket, min_support=0.5, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
print("Market basket association rules:\n", rules)

# Task 2: Medical symptoms and diseases (example)
# Synthetic example similar to market basket
symptoms = pd.DataFrame({
    'fever': [1,0,1,1],
    'cough': [1,1,0,1],
    'fatigue': [0,1,1,0],
    'flu': [1,0,1,1]
})
freq_symptoms = apriori(symptoms, min_support=0.5, use_colnames=True)
symptom_rules = association_rules(freq_symptoms, metric="lift", min_threshold=1.0)
print("Symptoms association rules:\n", symptom_rules)

# Task 3: Cooking ingredients associations
ingredients = pd.DataFrame({
    'salt': [1,1,0,1],
    'pepper': [1,0,1,1],
    'olive_oil': [0,1,1,1],
    'garlic': [1,1,1,0]
})
freq_ingredients = apriori(ingredients, min_support=0.5, use_colnames=True)
ingredient_rules = association_rules(freq_ingredients, metric="confidence", min_threshold=0.6)
print("Cooking ingredients association rules:\n", ingredient_rules)

```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

# -------------------------
# Customer Segmentation Examples
# -------------------------

# Task 1: Segmenting retail customers based on purchasing patterns (synthetic data)
retail_customers = np.array([
    [500, 20],   # [annual spending, visits per month]
    [1000, 30],
    [200, 10],
    [800, 25],
    [600, 15],
])
kmeans_retail = KMeans(n_clusters=2, random_state=0)
retail_clusters = kmeans_retail.fit_predict(retail_customers)
print("Retail customer segments:", retail_clusters)

# Task 2: Dividing social media users into groups based on engagement metrics
social_media_users = np.array([
    [100, 50],   # [likes, shares]
    [200, 80],
    [50, 20],
    [120, 60],
    [180, 70],
])
kmeans_social = KMeans(n_clusters=2, random_state=0)
social_clusters = kmeans_social.fit_predict(social_media_users)
print("Social media user segments:", social_clusters)

# Task 3: Grouping telecom users by usage patterns
telecom_users = np.array([
    [300, 400],  # [call minutes, data usage in MB]
    [500, 800],
    [200, 300],
    [450, 700],
    [350, 600],
])
kmeans_telecom = KMeans(n_clusters=2, random_state=0)
telecom_clusters = kmeans_telecom.fit_predict(telecom_users)
print("Telecom user segments:", telecom_clusters)


# -------------------------
# Anomaly Detection Examples
# -------------------------

# Task 1: Detect fraudulent credit card transactions (using Isolation Forest)
transactions = np.array([
    [100, 1], [120, 2], [130, 1], [5000, 20],  # 5000 is an outlier transaction amount
    [110, 1], [115, 2]
])
iso_forest = IsolationForest(contamination=0.1, random_state=0)
fraud_pred = iso_forest.fit_predict(transactions)
print("Fraud detection (-1 is anomaly):", fraud_pred)

# Task 2: Unusual network activity detection (simulated)
network_activity = np.array([
    [50, 20], [55, 18], [53, 19], [300, 100],  # Outlier
    [52, 21], [54, 20]
])
iso_forest_net = IsolationForest(contamination=0.1, random_state=0)
network_anomaly = iso_forest_net.fit_predict(network_activity)
print("Network activity anomaly (-1 is anomaly):", network_anomaly)

# Task 3: Monitoring industrial equipment for failure signs
equipment_sensors = np.array([
    [30, 0.5], [32, 0.6], [31, 0.55], [100, 5],  # Outlier sensor reading
    [29, 0.52], [33, 0.6]
])
iso_forest_eq = IsolationForest(contamination=0.1, random_state=0)
equipment_anomaly = iso_forest_eq.fit_predict(equipment_sensors)
print("Equipment anomaly detection (-1 is anomaly):", equipment_anomaly)


# -------------------------
# Feature Reduction Examples
# -------------------------

# Task 1: PCA on image dataset (using sklearn digits dataset)
from sklearn.datasets import load_digits

digits = load_digits()
pca = PCA(n_components=10)
digits_pca = pca.fit_transform(digits.data)
print("Explained variance ratios by top 10 components:", pca.explained_variance_ratio_)

# Task 2: Feature reduction for IoT sensor data (synthetic)
iot_data = np.random.rand(100, 50)  # 100 samples, 50 features
pca_iot = PCA(n_components=5)
iot_reduced = pca_iot.fit_transform(iot_data)
print("Reduced IoT sensor data shape:", iot_reduced.shape)

# Task 3: Simplifying marketing dataset by merging related variables (using PCA again)
marketing_data = np.random.rand(50, 10)
pca_marketing = PCA(n_components=3)
marketing_reduced = pca_marketing.fit_transform(marketing_data)
print("Reduced marketing data shape:", marketing_reduced.shape)

# Optional: Visualize PCA explained variance for digits dataset
plt.figure(figsize=(8,4))
plt.bar(range(1, 11), pca.explained_variance_ratio_, alpha=0.7)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('PCA Explained Variance Ratio (Digits Dataset)')
plt.show()
```

```python
# Difference Between Supervised & Unsupervised Learning

from sklearn.datasets import load_digits, fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Task 1: Supervised Learning - Predicting house prices based on labeled data
housing = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print("Task 1: Supervised regression predictions (house prices):", predictions[:5])


# Task 2: Unsupervised Learning - Exploring hidden demographics in customer profiles
# For simplicity, create synthetic data (e.g., age, income, spending score)
import numpy as np
customer_profiles = np.array([
    [25, 50000, 70],
    [40, 60000, 50],
    [22, 48000, 80],
    [38, 62000, 40],
    [26, 52000, 75],
])
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(customer_profiles)
print("Task 2: Customer profile clusters:", clusters)


# Task 3: Supervised Learning - Classifying emails as spam or not spam
# Use digits dataset as example classification dataset (digits 0-9)
digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, random_state=42)
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Task 3: Supervised classification accuracy:", accuracy_score(y_test, y_pred))


# Task 4: Unsupervised Learning - Grouping images of handwritten digits without labels
kmeans_digits = KMeans(n_clusters=10, random_state=42)
clusters_digits = kmeans_digits.fit_predict(digits.data)
print("Task 4: Unsupervised clustering of digits:", clusters_digits[:20])
```

```python
# Unsupervised Learning Examples

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Task 1: Identifying patterns in customer purchasing behavior (Clustering example)
# Sample synthetic customer data: features like purchase frequency, amount spent, categories bought
customer_data = np.array([
    [5, 200, 1],
    [3, 150, 0],
    [10, 600, 2],
    [7, 300, 1],
    [2, 100, 0]
])
kmeans = KMeans(n_clusters=2, random_state=42)
clusters = kmeans.fit_predict(customer_data)
print("Customer clusters:", clusters)


# Task 2: Grouping news articles by topic (Text clustering example)
documents = [
    "The economy is improving with higher jobs numbers",
    "Sports events are gaining more viewers",
    "Financial markets see a rise in stocks",
    "Local team wins championship game",
    "Investors optimistic about market growth"
]

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Cluster articles into 2 groups
kmeans_text = KMeans(n_clusters=2, random_state=42)
clusters_text = kmeans_text.fit_predict(X)
print("News article clusters:", clusters_text)


# Task 3: Analyzing satellite images to identify land use (Dimensionality reduction example)
# For illustration, create random high-dimensional data to simulate image features
image_features = np.random.rand(10, 100)  # 10 images, 100 features each

# Reduce dimensionality with PCA to 2 components for visualization
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(image_features)
print("Reduced image features (PCA):", reduced_features)
```
